```python
import pandas as pd
import numpy as np

#creating a simple m-pesa transactions dataset
data ={
    "name":["Alice", "bob", "Alice", "Carol", "Bob", "Alice"],
    "county": ["Nairobi", "Mombasa", "Nairobi", "kisumu", "Mombasa", "Nairobi"],
    "amount":[500, 1200, 300, 800, 450, 950],
    "type": ["send", "receive", "send", "send", "receive", "receive"]
}

df= pd.DataFrame(data)
print(df)
```

```python
#Filter - only "send" transactions
sends = df[df["type"]=="send"]
print(sends)
```

```python
#Group by county - total amount per county
by_county = df.groupby("county")["amount"].sum()
print(by_county)
```

```python
#sort - who received/sent the most money overall
by_person = df.groupby("name")["amount"].sum().sort_values(ascending= False)
print(by_person)
```

```python
# Clean the data — fix capitalization issues
df["name"] = df["name"].str.title()
df["county"] = df["county"].str.title()

# Now run the groupby again
by_person = df.groupby("name")["amount"].sum().sort_values(ascending=False)
print(by_person)
```

```python
df["is_large"] = df["amount"] > 600
print(df)
```
